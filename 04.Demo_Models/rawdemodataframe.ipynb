```python
def random_dates(start, end, n, unit='D', seed=None):
    if not seed:  
        np.random.seed(0)

    ndays = (end - start).days + 1
    return start + pd.to_timedelta(
        np.random.randint(0, ndays, n), unit=unit)

start = pd.to_datetime('1995-01-01')
end = pd.to_datetime('2002-01-01')
rdv = random_dates(start, end, 21)
rdv = rdv.to_frame(index=False, name='date_of_birth') #can replace observation date/time as needed too, as per next cell
```
