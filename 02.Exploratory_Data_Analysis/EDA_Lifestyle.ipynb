## Interactions

### Chi-square and heat map of interactions between all categorical variables

```python
from scipy.stats import chi2_contingency

# List of categorical variables
categorical_vars = ['diabetes', 'drinking_frequency', 'six_or_more_drinks_occurrence', 'smoking_status', 'drug_use', 'mental_health', 'birth_order']

# Create an empty DataFrame to store results
results_df = pd.DataFrame(columns=['Variable 1', 'Variable 2', 'Chi-square', 'P-value'])

# Loop through all possible pairs of categorical variables
for i in range(len(categorical_vars)):
    for j in range(i + 1, len(categorical_vars)):
        var1 = categorical_vars[i]
        var2 = categorical_vars[j]
        response = 'birth_class'
        
        contingency_table = pd.crosstab(index=[birth[var1], birth[var2]],columns=birth[response])
        sns.heatmap(contingency_table, annot=True, cmap='viridis', fmt='d')
        plt.show()

        # Perform the chi-square test
        chi2, p, _, _ = chi2_contingency(contingency_table)
        print(f'Interactions between: {var1} and {var2}')
        print(f'Chi-square value: {chi2}')
        print(f'P-value: {p}')
        
        # Append results to DataFrame
        results_df = results_df.append({'Variable 1': var1, 'Variable 2': var2, 'Chi-square': chi2, 'P-value': p}, ignore_index=True)
```

## Stacked bar plot of preterm births by diabetes status

```python
# calculate percentage of term vs. preterm by category, then make stacked bar plot

grouped_diabetes = birth.groupby('diabetes')['birth_class'].value_counts(normalize=True).unstack().fillna(0) * 100

# Define colors for 'Term' and 'Preterm'
term_color = 'green'  
preterm_color = 'orange'

category_order = ['Type1', 'Type2', 'TypeUnknown','None']

# Create a stacked bar plot
fig, ax = plt.subplots(figsize=(10, 6), dpi=300)
ax = grouped_diabetes.loc[category_order].plot(kind='bar', stacked = True, figsize=(10, 6),color=[preterm_color, term_color], edgecolor='black', ax=ax)

# Change category labels
new_labels = ['Type 1', 'Type 2', 'Type Unknown', 'None']
ax.set_xticklabels(new_labels, rotation=0, fontsize=12)

# Edit legend
legend_title = 'Birth Class'
ax.legend(title=legend_title, title_fontsize=12, loc='upper left', bbox_to_anchor=(1, 1),fontsize=12)

# Set labels and title
plt.xlabel('Diabetes Status',fontsize=14)
plt.ylabel('Percentage (%)',fontsize=14)
plt.title('Percentage of Term vs. Preterm Births by Diabetes Status',fontsize=16)
ax.tick_params(axis='y', labelsize=12)
```

## Heat map of interaction between drinking and drug use

```python
# Define the order you want for the drinking frequency categories
drinking_frequency_order = ['4orMoreWeekly', '2to3Weekly', '2to4Monthly','Monthly','Skipped']
drug_use_order = ['Yes','No']

# Create a new categorical column with the desired order
birth['drinking_frequency_ordered'] = pd.Categorical(birth['drinking_frequency'], categories=drinking_frequency_order, ordered=True)
birth['drug_use_ordered'] = pd.Categorical(birth['drug_use'], categories=drug_use_order, ordered=True)

var1 = 'drinking_frequency_ordered'
var2 = 'drug_use_ordered'
response = 'birth_class_binary'

pivot_table = birth.pivot_table(index=var1, columns=var2, values=response, aggfunc='mean', fill_value=0)

# Create a high-resolution plot
fig, ax = plt.subplots(figsize=(5, 5), dpi=300)
ax = sns.heatmap(pivot_table, annot=True, cmap='Oranges', fmt='.2%', cbar_kws={'label': 'Proportion of Preterm Births'},annot_kws={"size": 12})

# Adjust color bar label size, rotation, and position
cbar = ax.collections[0].colorbar
cbar.set_label('Proportion of Preterm Births', size=14)
cbar.ax.set_yticklabels(cbar.ax.get_yticklabels(), fontsize=10,rotation=90)
#cbar_ticks = [0.0, 0.05, 0.1, 0.15, 0.2, 0.25, 0.35, 0.40]  # Specify the desired tick positions
#cbar.set_ticks(cbar_ticks)
#cbar.ax.set_yticklabels(cbar.ax.get_yticklabels(), fontsize=12,rotation=90,fmt='.2%')

# Set axis titles and labels
ax.set_xlabel('Drug Use',fontsize=14)
ax.set_ylabel('Drinking Frequency',fontsize=14)
ax.set_xticklabels(['Yes','No'],rotation=0,fontsize=12)

# Change y-axis labels
new_y_labels = ['4+ Weekly', '2-3 Weekly', '2-4 Monthly','Monthly','Skipped']
ax.set_yticklabels(new_y_labels, rotation=0,fontsize=12)

# Title and show the plot
plt.title('Drinking Frequency & Drug Use\n Increase Proportion of Preterm Births',fontsize=16)
plt.show()
```
