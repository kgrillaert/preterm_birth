## Categorical Data

### Race count plot

### Race stacked bar plot

```python
# calculate percentage of term vs. preterm by category, then make stacked bar plot

grouped_race = birth.groupby('race_person')['birth_class'].value_counts(normalize=True).unstack().fillna(0) * 100

# Define colors for 'Term' and 'Preterm'
term_color = '#1f77b4'  # Blue
preterm_color = '#ff7f0e'  # Orange

category_order = ['White','Black or African American', 'Asian','More than one population','Middle Eastern or North African','Native Hawaiian or Other Pacific Islander','None of these', 'no answer']

# Create a stacked bar plot
ax = grouped_race.loc[category_order].plot(kind='bar', stacked=True, figsize=(8, 6), color=[preterm_color, term_color])

# Set labels and title
plt.xlabel('Race')
plt.ylabel('Percentage')
plt.title('Percentage of Term vs. Preterm Births by Race')
```

# Plots for Presentation

## Percentage of preterm births by race (bar plot)

```python
# calculate percentage of term vs. preterm by category, then make stacked bar plot

grouped_race = birth.groupby('race_person')['birth_class'].value_counts(normalize=True).unstack().fillna(0) * 100

# Define colors for 'Term' and 'Preterm'
term_color = 'green'  
preterm_color = 'orange'

category_order = ['White','Black or African American', 'Asian','Middle Eastern or North African','Native Hawaiian or Other Pacific Islander','More than one population','None of these', 'no answer']

# Create a stacked bar plot
fig, ax = plt.subplots(figsize=(10, 6), dpi=300)
ax = grouped_race.loc[category_order].plot(kind='bar', stacked = True, figsize=(10, 6),color=[preterm_color, term_color], edgecolor='black', ax=ax)

# Change category labels
new_labels = ['White', 'Black/\nAfrican\nAmerican', 'Asian', 'Middle\nEastern/\nNorth\nAfrican','Native\nHawaiian/\nPacific\nIslander', 'More than\none','None\nof these', 'No answer']
ax.set_xticklabels(new_labels, rotation=0, fontsize=12)

# Edit legend
legend_title = 'Birth Class'
ax.legend(title=legend_title, title_fontsize=12, loc='upper left', bbox_to_anchor=(1, 1),fontsize=12)

# Set labels and title
plt.xlabel('Race',fontsize=14)
plt.ylabel('Percentage (%)',fontsize=14)
plt.title('Percentage of Term vs. Preterm Births by Race',fontsize=16)
ax.tick_params(axis='y', labelsize=12)
```

## Interaction of age & income with birth class

```python
# Set the figure size and higher dpi
plt.figure(figsize=(6, 4))

custom_palette = {'Term': 'green', 'Preterm': 'orange'}

lm = sns.lmplot(data = birth,
                x = "age_at_birth",
                y = 'median_income_zip',
                hue = 'birth_class',
                palette=custom_palette,
                scatter = False,
                line_kws={'linewidth': 3})

# Remove the default legend created by lmplot
lm.fig.get_children()[-1].remove()

# Access the Matplotlib figure and set higher dpi
lm.fig.set_dpi(300)

# Set labels and title
plt.xlabel('Maternal Age (yrs)', fontsize=14)
plt.ylabel('Median Income in Zip Code ($)', fontsize=14)
plt.title('Relationships between Maternal Age & Income\nby Birth Class', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.show()
```
