## Continuous Data

### Pairplot

```python
plt.figure(figsize=(12,12))

cols_to_plot = ['birth_class', 'age_at_birth', 'birth_order', \
                'assisted_income_zip','high_school_education_zip','median_income_zip', \
                'no_health_insurance_zip','poverty_zip','vacant_housing_zip','deprivation_index_zip']

sns.pairplot(birth[cols_to_plot], hue='birth_class')

plt.show()
```

### Histogram of mother's age by birth class

```python
plt.figure(figsize=(10,4))

sns.set_style('whitegrid')
sns.kdeplot(data=birth,
            x='age_at_birth', 
            hue='birth_class', 
            fill=True,
            common_norm=False)

plt.xlabel("Mother's Age at Conception (yrs)")
plt.ylabel('Count')
plt.show()
```

## Categorical Data

### Race count plot

```python
plt.figure(figsize=(16,4))

sns.histplot(data = birth,
            x='race_person',
            hue='birth_class',
            multiple='dodge', shrink=0.75,
            stat='proportion')

plt.xticks(rotation=30)

plt.show()
```

## Interactions

### Box plot for race & median income by birth class

```python
plt.figure(figsize=(14,6))

sns.boxplot(data = birth,
             x = 'race_person',
             y = 'median_income_zip',
             hue = 'birth_class')

plt.xlabel("Race")
plt.ylabel("Median Income ($)")
plt.xticks(rotation=30)
plt.show()
```

### Box plot for ethnicity & median income by birth class

```python
plt.figure(figsize=(14,6))

sns.boxplot(data = birth,
              x = 'ethnicity_person',
              y = 'median_income_zip',
              hue = 'birth_class')

plt.xlabel("Ethnicity")
plt.ylabel("Median Income ($)")
plt.xticks(rotation=30)
plt.show()
```

### Box plot for race & education by birth class

```python
plt.figure(figsize=(14,6))

sns.boxplot(data = birth,
            x = 'race_person',
            y = 'high_school_education_zip',
            hue = 'birth_class')

plt.xlabel("Race")
plt.ylabel("% of people in zipcode who completed high school")
plt.xticks(rotation=30)
plt.show()
```

### Box plot for ethnicity & education by birth class

```python
plt.figure(figsize=(14,6))

sns.boxplot(data = birth,
            x = 'ethnicity_person',
            y = 'high_school_education_zip',
            hue = 'birth_class')

plt.xlabel("Ethnicity")
plt.ylabel("% of people in zipcode who completed high school")
plt.xticks(rotation=30)
plt.show()
```

### Scatter plot for income & age by birth class

```python
plt.figure(figsize=(14,6))

sns.lmplot(data = birth,
            x = "age_at_birth",
            y = 'median_income_zip',
            hue = 'birth_class',
            scatter_kws={'alpha':0.05})

plt.xlabel("Age")
plt.ylabel("Median Income in Zip Code")
plt.xticks(rotation=30)
plt.show()
```
